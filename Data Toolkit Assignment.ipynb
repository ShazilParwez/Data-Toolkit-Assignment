3.Explain the following terms:
  
  The difference in np.array, np.asarray and np.asanyarray.
  
  The difference between Deep copy and shallow copy.

Answer 3:

#NumPy Array Functions

1. `np.array`:
   - This function creates a new NumPy array from any input data, such as lists or tuples. Importantly, it always makes a copy of the data. If you use `np.array`, you can be sure that the new array is separate from the original input, even if that input is already an array.

2. `np.asarray`:
   - This function also creates a NumPy array, but with a key difference: if the input is already a NumPy array, `np.asarray` will return it without making a copy. It only creates a new array when the input is a different type, like a list. This makes it more efficient when you want to convert data but avoid unnecessary copying.

3. `np.asanyarray`:
   - This function is similar to `np.asarray`, but it goes a step further by allowing for the return of array subclasses (like matrices or masked arrays) without converting them to a base array type. This means that if you provide it with a special type of array, it will return that same type rather than changing it to a standard array.

#Deep Copy vs. Shallow Copy

1. Shallow Copy:
   - A shallow copy creates a new object but keeps references to the original nested objects. This means if you modify one of the nested objects in the shallow copy, it will also affect the original. In essence, while the outer structure is new, the inner data is shared.

2. Deep Copy:
   - A deep copy creates a completely independent object, along with copies of all objects contained within it, no matter how deeply nested. Changes made to a deep copy do not affect the original object at all. This is useful when you need to ensure that the two objects operate independently from each other.

```python
#5.Create a Numpy array with random integers between 1 and 10 of shape (5, 6). After creating the array
#perform the following operations:
  #a)Extract all even integers from array.
  #b)Extract all odd integers from array.

#Answer 5:

import numpy as np

# Create a 5x6 array with random integers between 1 and 10
array_random = np.random.randint(1, 11, size=(5, 6))

print("Original Array:")
print(array_random)

# a) Extract all even integers from the array
even_integers = array_random[array_random % 2 == 0]
print("\nEven Integers:")
print(even_integers)

# b) Extract all odd integers from the array
odd_integers = array_random[array_random % 2 != 0]
print("\nOdd Integers:")
print(odd_integers)
```

```python
#15.Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
  #a)Plot a histogram of the data with 30 bins.
  #b)Overlay a line plot representing the normal distribution's probability density function (PDF).
  #c)Label the x-axis as Value and the y-axis as Frequency/Probability.
  #d)Set the title of the plot as Histogram with PDF Overlay.

#Answer 15:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # mean=0, std=1

# a) Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# b) Overlay a line plot representing the normal distribution's PDF
# Create an array of x values for the PDF
x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, loc=0, scale=1)  # Calculate the PDF for the normal distribution

plt.plot(x, pdf, color='red', label='PDF', linewidth=2)

# c) Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# d) Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
#16.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plotâ€™.

#Answer 16:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 10  # Random values between 0 and 10
y = np.random.rand(100) * 10  # Random values between 0 and 10

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Determine the quadrant
def get_quadrant(row):
    if row['X'] >= 5 and row['Y'] >= 5:
        return 'Quadrant 1'
    elif row['X'] < 5 and row['Y'] >= 5:
        return 'Quadrant 2'
    elif row['X'] < 5 and row['Y'] < 5:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='viridis', style='Quadrant', markers=['o', 's', 'D', 'X'])

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title of the plot
plt.title('Quadrant-wise Scatter Plot')

# Show legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

```python
#17.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

#Answer 17:

from bokeh.plotting import figure, show
import numpy as np

# Generate data for the sine wave
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)',
           plot_width=800, plot_height=400)

# Add a line renderer
p.line(x, y, legend_label='sin(x)', line_width=2, line_color='blue')

# Add grid lines
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```
